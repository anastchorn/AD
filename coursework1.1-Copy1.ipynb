10) Ви можете вставляти 3D об'єкти у Jupyter Notebook, зокрема, використовуючи бібліотеку pythreejs. Ця бібліотека дозволяє створювати та відображати 3D сцени у Jupyter Notebook.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Генерація даних
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, label='3D curve')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Plot')
ax.legend()
plt.show()
```

4) Відобразити 10 елементів для 4-ох ознак Виділити зеленим кольором максимальні елементи в таблиці, а червоним мінімальні для цих 10-ти елементів.

```python
selected_data = df[['mpg', 'cylinders','displacement', 'horsepower']].head(10)
max_val = selected_data.values.max()
min_val = selected_data.values.min()

# Функція для підсвічення
def highlight_max_min(val):
    if val == max_val:
        return 'background-color: green'
    elif val == min_val:
        return 'background-color: red'
    else:
        return ''

# Застосування функції до кожного елемента датафрейму за дпомогою методу Style
styled_result_matrix = selected_data.style.map(highlight_max_min)
styled_result_matrix
```

5) Вирахувати математичне сподівання та дисперсію для набору даних. Стандартизувати дані.

**Математичне сподівання** (середнє значення або очікуване значення) - це поняття в теорії ймовірностей та статистиці, яке вказує на середнє значення випадкової величини

$$X_{\text{std}} = \frac{x - \mu}{\sigma}$$

$\mu = \sum p_i x_i$


$$\sigma = \sqrt{\frac{\sum(x_i - \bar{x})}{n-1}}$$


```python
# Mat Expectation
non_numeric_columns = ['name', 'origin']
df_matexp = df.drop(non_numeric_columns, axis=1)

mean_values = df_matexp.mean()
mean_values
```

**Дисперсія** - це міра розсіювання значень відносно їх середнього значення в наборі даних. Дисперсія дозволяє визначити, наскільки далеко кожне конкретне значення відхиляється від середнього значення.

```python
# dispersion
dispersion = df.var(numeric_only=True)
dispersion
```

**Стандартизація даних** - це процес перетворення значень змінних так, щоб вони мали стандартний розподіл з середнім значенням 0 і стандартним відхиленням 1. Цей процес часто включає в себе віднімання середнього значення від кожного значення та поділ на стандартне відхилення.

```python
#standard data
numeric_columns = df.select_dtypes(include=[float, int])
df_standart=df.copy()
df_standart[numeric_columns.columns] = ((numeric_columns - numeric_columns.mean()) / numeric_columns.std()).round(2)
print("standard data:")
print(df_standart)
```

11) Видалити також випадкові значення з ознак що мають найбільшу кореляцію у кількості 10% для обох. Підрахувати кореляцію. Збудувати multipple scatterplot за допомогою FacetGrid для цих оновлених ознак використовуючи бібліотеку seaborn взявши третю ознаку за вісь x.

**sns.FacetGrid** - це об'єкт бібліотеки Seaborn для створення графіків з використанням концепції "facet" або "facetted" графіків. Графіки з "facet" розділяють дані на підгрупи відповідно до значень одного чи кількох змінних, і створюють окремий графік для кожної підгрупи.

```python
#відсоток рядків, які потрібно видалити
percentage_to_remove = 0.10
#обчислюємо кількість рядків, які потрібно видалити
rows_to_remove = int(percentage_to_remove * len(df))
# індекси рядків, які потрібно видалити
random_rows_index = np.random.choice(df.index, size=rows_to_remove, replace=False)

df_col.loc[random_rows_index, ['displacement', 'cylinders']] = np.nan
print("Кількість видаленних випадкових значень:")
print(df_col.isnull().sum())
print("Датасет після видалення випадкових значень:")
print(df_col)
df_col =df_col.ffill()
print(df_col)
print("Кореляція після видалення випадкових значень:")
print(df_col.corr(method="pearson", numeric_only=True))

# Побудова multiple scatterplot за допомогою FacetGrid
g = sns.FacetGrid(df_col[(df_col['origin'] == "usa") | (df_col['origin'] == "europe")], col='origin', hue='cylinders', height=4)
g.map(sns.scatterplot, 'weight', 'displacement')
g.add_legend()
plt.show()
```
